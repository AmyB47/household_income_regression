```python
import pickle
import pandas as pd
import numpy as np
```

```python
with open('county_data.pickle', 'rb') as read_file:
    county_data = pickle.load(read_file)
    
county_data.head()
```

```python
len(county_data)
```

```python
cnty_data = county_data[['fips_txt',
                        'FIPS_code_NEW',  #ST FIPS + County FIPS
                       'Stabr',  # ST abbreviation
                       'area_name',   # county, ST
                       'Area_Name',   # county
                         
                       'Median_Household_Income_2019',  # target
                         
                       'POP_ESTIMATE_2019', 
                         
                       'Civilian_labor_force_2019',
                       'Employed_2019', 
                       'Unemployed_2019',
                       'Unemployment_rate_2019', 

                       'Metro_2013',
                       'Rural_urban_continuum_code_2013',   
           
                       'Farming_2015_Update', 
                       'Mining_2015-Update',
                       'Manufacturing_2015_Update', 
                       'Government_2015_Update',
                       'Recreation_2015_Update', 
                       'Nonspecialized_2015_Update',
                         
                       'Low_Education_2015_Update', 
                       'Low_Employment_Cnty_2008_2012_25_64', 
                        
                        'Retirement_Dest_2015_Update',
                        'Persistent_Poverty_2013', 
                         
                        'hwy_count',
                        'city_count',
                        'town_count' ]]
```

```python
cnty_data.head()
```

```python
# review for any null 

cnty_data.info()
```

```python
cnty_data[cnty_data['FIPS_code_NEW'].isna()]
```

```python
cnty_data['FIPS_code'] = cnty_data['fips_txt'].apply(lambda x: str(x).zfill(5))
```

```python
cnty_data[cnty_data['hwy_count'].isna()]
```

```python
# one of the data sources contained census areas for AK in addition to counties
# Also VA independent cities 
# add in 0 for DC hwy, towns, cities

# research these features for the additional areas of VA and AK time-permitting

cnty_data.loc[cnty_data['Stabr'] == 'DC', ['hwy_count', 'city_count', 'town_count']] = 0
cnty_data.loc[cnty_data['Stabr'] == 'DC', ['FIPS_code_NEW']] = '11001'

cnty_data.loc[cnty_data['Stabr'] == 'DC']
```

```python
model_data = cnty_data.dropna()

model_data.info()
```

```python
# pop_estimate_2019 needs converted to number
model_data['pop_est_2019'] = model_data['POP_ESTIMATE_2019'].str.replace(',','').astype(float)

```

```python
# drop columns not needed 
model_data.drop(['fips_txt', 'FIPS_code_NEW', 'POP_ESTIMATE_2019'], axis=1, inplace=True)
```

```python
model_data.columns
```

```python
# reorder columns for easy splitting into ranges for eda and model

eda_ready = model_data[[ 
                        # county/area identifiers
                        'FIPS_code','Stabr', 'area_name', 'Area_Name', 
                        
                        # target
                        'Median_Household_Income_2019', 
                        
                        # numeric features
                        'pop_est_2019', 'Civilian_labor_force_2019', 'Employed_2019', 
                        'Unemployed_2019', 'Unemployment_rate_2019', 
                        'hwy_count', 'city_count', 'town_count',
                        
                        # category features
                       'Metro_2013',
                       'Rural_urban_continuum_code_2013', 'Farming_2015_Update',
                       'Mining_2015-Update', 'Manufacturing_2015_Update',
                       'Government_2015_Update', 'Recreation_2015_Update',
                       'Nonspecialized_2015_Update', 'Low_Education_2015_Update',
                       'Low_Employment_Cnty_2008_2012_25_64', 'Retirement_Dest_2015_Update',
                       'Persistent_Poverty_2013' ]]
```
